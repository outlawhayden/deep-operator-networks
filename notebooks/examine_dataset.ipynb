```python
import numpy as np

data = np.load("../data/pde_source_solution_dataset.npy", allow_pickle = True).item()

x = data["x"]
t = data["t"]
sources = data["sources"]
solutions = data["solutions"]
```

```python
solutions.shape
```

```python
import matplotlib.pyplot as plt
plt.plot(x, sources[0])
```

```python
X, T = np.meshgrid(x, t)   # make 2D grids

plt.figure(figsize=(7,5))
plt.pcolormesh(X, T, solutions[0], shading="auto", cmap="viridis")
plt.gca().invert_yaxis() 
plt.colorbar(label="s(x,t)")
plt.xlabel("x")
plt.ylabel("t")
plt.title("s(x,t), samp 0")
plt.show()
```
