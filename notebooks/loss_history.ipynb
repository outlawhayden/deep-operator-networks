```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
branch_losses = np.load("/Users/haydenoutlaw/Documents/Research/opnet/deep-operator-networks/deeponet_2step/deeponet_branch_losses.npz")
br_test= branch_losses['test']
br_epochs = np.arange(1,br_test.shape[0] + 1)

br_test
```

```python
trunk_losses = np.load("/Users/haydenoutlaw/Documents/Research/opnet/deep-operator-networks/deeponet_2step/deeponet_trunk_losses.npz")
tr_test= trunk_losses['test']
tr_epochs = np.arange(1, tr_test.shape[0] + 1)
tr_test
```
