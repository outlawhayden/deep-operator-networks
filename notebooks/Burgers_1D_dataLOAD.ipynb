```python
import matplotlib.pyplot as plt
import numpy as np

idx_data = 3

x_grid = np.linspace(-3,3,201)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(x_grid, DATA['data'][idx_data]['x'][0,:],label=f't={DATA['data'][0]['t'][0].item()}')
axes[0].plot(x_grid, DATA['data'][idx_data]['x'][-1,:],label=f't={DATA['data'][0]['t'][-1].item()}')
axes[0].set_xlabel('x')
axes[0].set_ylabel('u (solution)')
axes[0].set_title(f'alpha= {DATA['param'][idx_data][0]} and omega= {DATA['param'][idx_data][1]}')
axes[0].set_xlim([-3.1, 3.1])
axes[0].set_ylim([-0.1, 0.8])
axes[0].legend()
axes[1].plot(x_grid, DATA['data'][idx_data]['dx'][0,:],label=f't={DATA['data'][idx_data]['t'][0].item()}')
axes[1].plot(x_grid, DATA['data'][idx_data]['dx'][-1,:],label=f't={DATA['data'][idx_data]['t'][-1].item()}')
axes[1].set_xlabel('x')
axes[1].set_ylabel('du/dx')
axes[1].set_title(f'alpha= {DATA['param'][idx_data][0]} and omega= {DATA['param'][idx_data][1]}')
axes[1].set_xlim([-3.1, 3.1])
axes[1].set_ylim([-0.5, 2.0])
axes[1].legend()
plt.show()
```
